## Preparing the tools 
* We are going to use pandas for data processing, numpy for linear algebra, matplotlib and seaborn for data visualization.

```python
# Importing all the necessary tools

# Regular EDA and plotting libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importing the Dataset

```python
df = pd.read_csv("Ecommerce Purchases")
```

### What is the peak time of purchase

```python
df['AM or PM'].value_counts().plot(kind = "bar", color = ["teal","darkorange"],figsize = (10,6));
```

### Which browser do people use the most

```python
df['Browser Info'] = df['Browser Info'].str.split("/",expand = True).iloc[:,[0]]
df['Browser Info'].value_counts().plot(kind = "pie", autopct='%1.0f%%',colors = ["teal","darkorange"],figsize = (10,6),ylabel ='' );
```

### Statistics of CC Provider

```python
df['CC Provider'].value_counts().plot(kind = 'pie', autopct='%1.0f%%',figsize = (10,6),ylabel ='');
```

### Maximum purchase 

```python
df['Purchase Price'].max()
```

### Minimum Purchase

```python
df['Purchase Price'].min()
```

### Engineering Jobs

```python
df[df['Job'].str.contains('engineer')]
```

### Mean Purchase price

```python
df['Purchase Price'].mean()
```
